```python
#del df['Tax']
df.insert(6,'Tax','0.00')
```

```python
def set_tax(column):
    if column['Income'] in range(10000,40000):
        return 0.15
    elif column['Income'] in range(40000,80000):
        return 0.20
    else:
        return 0.25

df = df.assign(Tax = df.apply(set_tax, axis=1))
df
```
